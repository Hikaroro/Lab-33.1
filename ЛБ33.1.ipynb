2. вторичная цепь, содержит измерительную катушку $L{\scriptscriptstyle И}$ и баллистический гальванометр  $Г$.

### Дайте определение физических величин: магнитный момент тока, магнитная индукция, напряжённость магнитного поля, магнитный поток. Каковы единицы измерения этих величин?

    
Магнитный момент тока $\overrightarrow{p}{\scriptscriptstyle m}$ - физическая величина, характеризующая контур с током, численно равен произведению силы тока $I$ в контуре на площадь поверхности контура (рамки) $S$ и направлен вдоль нормали $\overrightarrow{n}$ 
$\overrightarrow{p}{\scriptscriptstyle m} = IS\overrightarrow{n}$. Единица измерения - $A*м^2$
    
Вектор магнитной индукции $\overrightarrow{B}$  является количественной силовой характеристикой магнитного поля. Магнитная индукция в данной точке магнитного поля численно равна максимальному вращающему моменту, действующему на рамку с током, магнитный момент которой равен единице: $B = {M{\scriptscriptstyle max} \over p{\scriptscriptstyle m}}$. Единицей измерения магнитной индукции в СИ является Тесла (Тл).
    
Напряженность магнитного поля $\overrightarrow{H}$ называют векторную величину, которая в случае однородной изотропной среды связана с индукцией магнитного поля соотношением $\overrightarrow{B} = \mu \mu{\scriptscriptstyle 0}\overrightarrow{H}$. Единицей измерения напряженности поля в СИ является 1 *А/м* (Ампер на метр).
    
Потоком вектора индукции (магнитным потоком) через поверхность $S$ называется скалярная величина $\Phi = \int_{s}B{\scriptscriptstyle n}dS = \int_{s}(\overrightarrow{B}  d\overrightarrow{S}) = \int_{S}B\cos\alpha dS$, где $B{\scriptscriptstyle n}$ **- проекция вектора $B$ на нормаль $n$ **к площадке $dS$ **, взятой на поверхности $S$ **; $d\overrightarrow{S} =  \overrightarrow{n}dS$*.* Поток $\Phi$ равен (пропорционален) числу линий индукции, пронизывающих данную поверхность $S$, при условии, что $B$ **перпендикулярен к $S$ **. Единицей магнитного потока в СИ является Вебер (Вб).

### В чём заключается метод измерения индукции магнитного поля с помощью баллистического гальванометра? Почему величина индуцированного заряда не зависит от скорости изменения магнитного потока через измерительную катушку.

В рассматриваемую точку исследуемого магнитного поля перпендикулярно линиям индукции вносится измерительная катушка, имеющая $N$ витков и площадь каждого витка $S$. Размеры измерительной катушки делаются достаточно малыми, чтобы при измерениях приблизиться, по возможности, к значениям индукции в данной точке поля. Катушку замыкают на баллистическом гальванометре $Г$. Магнитный поток через катушку будет равен $\Phi = BNS$
. 

При изменении магнитного потока в цепи гальванометра будет наводиться э.д.с. индукции. Если изменить направление вектора индукции поля на противоположное (это достигается изменением направления тока, создающего это поле), то изменение магнитного потока через катушку будет  $\Phi {\scriptscriptstyle 1} -\Phi {\scriptscriptstyle 2} = \Phi - (-\Phi) = 2\Phi$
 , и на основании  $Q = \int_{0}^{\tau}I{\scriptscriptstyle i}dt = -\int_{\Phi{\scriptscriptstyle 1}}^{\Phi{\scriptscriptstyle 2}}{1 \over R}d\Phi = {\Phi{\scriptscriptstyle 1} - \Phi{\scriptscriptstyle 2} \over R}$
 (Наглядно видно, что заряд $Q$ не зависит от скорости изменения потока)  в катушке будет индуцирован заряд $Q = {2\Phi \over R} = {2BNS \over R}$, где $R$  **– сопротивление цепи (катушка и гальванометр). Величина   ${2NS \over R}$ является постоянной для данной измерительной катушки.

Заряд, индуцированный в катушке, измеряется баллистическим гальванометром и равен $Q = km$, где $k$ **– постоянная гальванометра (цена деления его шкалы), измеряется в Кулонах на деление шкалы; $m$ **– число делений по шкале гальванометра, на которое отклоняется его стрелка (или световой «зайчик»). Используя $Q = {2\Phi \over R} = {2BNS \over R}$ и $Q = km$ можно найти индукцию магнитного поля в центре измерительной катушки, т.е. индукцию исследуемого поля $B = {kmR \over 2NS } = akm$, где $a$ - постоянная для данной установки величина. Значения $a$ и постоянной гальванометра $k$ указаны в паспорте установки.

### Сформулируйте и запишите закон Био-Савара-Лапласа. В чём его значение? Как он используется в данной работе?

Закон Био-Савара-Лапласа: $d\overrightarrow{B}= {\mu \mu{\scriptscriptstyle 0} \over 4\pi}{I [d\overrightarrow{l}, \overrightarrow{r}] \over r^3}$, где $\overrightarrow{r}$ радиус-вектор, проведенный от элемента $d\overrightarrow{l}$ проводника в точку А, $\mu$ - магнитная проницаемость среды, окружающей проводник с током (для вакуума $\mu = 1$), $\mu{\scriptscriptstyle 0} = 4\pi * 10^{-7} {Гн\overм}$ - магнитная постоянная. Вектор $d\overrightarrow{B}$ лежит в плоскости, перпендикулярной элементу тока $d\overrightarrow{l}$.

Закон  Био-Савара-Лапласа используется для нахождения магнитной индукции поля, создаваемого всем током в рассматриваемой точке A. Для этого необходимо проинтегрировать соотношение $d\overrightarrow{B}= {\mu \mu{\scriptscriptstyle 0} \over 4\pi}{I [d\overrightarrow{l}, \overrightarrow{r}] \over r^3}$ вдоль всего проводника. При этом поля от различных элементов тока векторно складываются, т. е. используется принцип суперпозиции для магнитных полей $B = \int dB$.

В данной работе на основании закона Био-Савара-Лапласа и принципа суперпозиции выводится формула напряженности магнитного поля на оси кругового тока, которая используется для подсчета теоретических значений.

### Выведите формулу для индукции магнитного поля в центре кругового тока $I$ известного радиуса $R$.

На круговом проводнике, по которому течет постоянный ток  $I$, выделем эдемент  $dl$, который можно считать прямолинейным. если перейти к другому элементу этого же тока, затем к третьему и так далее, применить правило винта, то все магнитные поля, созданные этими элементами в центре направлены вдоль одной прямой, перпендикуляру к плоскости кольца. Это означает, применяя принцип суперпозиции, векторное сложение можно заменить алгебраическим.

Закон Био-Савара-Лапласа для модуля вектора индукции поля, создаваемого элементом  $dl{\scriptscriptstyle 1}$:  $B= {\mu \mu{\scriptscriptstyle 0} \over 4\pi}{I dl{\scriptscriptstyle 1}, \sin\alpha \over r^2}$.

Так как расстояние от элементарного тока до центра витка равно радиусу $R$ и будет одинаковым для всех элементов на этом витке и элементы $dl$ будут нормальны к радиус-вектору $\overrightarrow{r}$, то $B= {\mu \mu{\scriptscriptstyle 0} \over 4\pi}{I dl \over R^2}$. 

Для нахождения полного поля в центра тока проинтегрируем($L = 2\pi R$ - длина окружности витка):  $B= \oint_{L}dB ={\mu \mu{\scriptscriptstyle 0} \over 4\pi}{I \over R^2}\oint_{L}dl = {\mu \mu{\scriptscriptstyle 0} \over 4\pi}{I \over R^2}2\pi R$.

Получаем $B = \mu \mu{\scriptscriptstyle 0}{I \over 2R}$

### Сформулируйте и запишите закон электромагнитной индукции Фарадея. Как и почему применяется этот закон в данной работе? 

Явление электромагнитной индукции состоит в том, что в замкнутом проводящем контуре при изменении потока магнитной индукции через поверхность, охватываемую этим контуром, возникает электрический ток, получивший название индукционного.

Наличие этого тока указывает, что в контуре возникает электродвижущая сила индукции, которая численно равна и противоположна по знаку скорости изменения магнитного потока сквозь поверхность, ограниченную этим контуром $\epsilon{\scriptscriptstyle i} = -{d\Phi \over dt}$ .

Данный закон применяется для нахождения заряда $Q$, прошедшего в контуре при изменении магнитного потока, который необходим для подсчета индукции исследуемого поля:

Направление индукционного тока определяется правилом Ленца: индукционный ток направлен так, чтобы своим магнитным полем противодействовать изменению магнитного потока, которым он вызван.

На основании закона электромагнитной индукции можно найти количество электричества (заряд), индуцируемое в контуре при изменении магнитного потока. Если контур обладает сопротивлением *R* , то мгновенное значение силы индукционного тока в нем на основании закона электромагнитной индукции Фарадея и закона Ома равно $I{\scriptscriptstyle i} = {\epsilon{\scriptscriptstyle i} \over R} = {1 \over R}{d\Phi \over dt}$.

Заряд, прошедший в контуре при изменении магнитного потока от величины  $\Phi{\scriptscriptstyle 1}$ до $\Phi{\scriptscriptstyle 2}$ за время $\tau$, равен $Q = \int_{0}^{\tau}I{\scriptscriptstyle i}dt = -\int_{\Phi{\scriptscriptstyle 1}}^{\Phi{\scriptscriptstyle 2}}{1 \over R}d\Phi = {\Phi{\scriptscriptstyle 1} - \Phi{\scriptscriptstyle 2} \over R}$

### Сформулируйте теорему о циркуляции вектора  $\overrightarrow{H}$ и запишите ее выражение. Пользуясь этой теоремой, найдите напряженность магнитного поля в середине тонкого длинного соленоида и внутри тороида.

Теорема о циркуляции вектора $\overrightarrow{H}$: $\oint_{L}H{\scriptscriptstyle l} dl = \sum_{k} I{\scriptscriptstyle k}$ , где $\oint_{L}H{\scriptscriptstyle l} dl$ - циркуляция вектора $\overrightarrow{H}$ по произвольному контуру $L$, $\sum_{k} I{\scriptscriptstyle k}$ -алгебраическая сумма сил токов, охватываемых контуром, $H{\scriptscriptstyle l} = H\cos\alpha$ - проекция вектора $\overrightarrow{H}$ на направления вектора $d\overrightarrow{l}$ - элемента контура $L$. Угол $\alpha$ - угол между векторами $\overrightarrow{H}$ и $d\overrightarrow{l}$. 

При непрерывном распределении тока в среде $\oint_{L}H{\scriptscriptstyle l} dl = \int_{S}j{\scriptscriptstyle n}dS$ , где $S$ - поверхность, ограниченная контуром  $L$, а $j{\scriptscriptstyle n}$ - проекция вектора плотности макроскопического тока на нормаль $\overrightarrow{n}$ к $S$.

Для нахождения напряженность магнитного поля в середине тонкого длинного соленоида воспользуемся теоремой о циркуляции вектора $\overrightarrow{H}$:  $\oint H dl = nl$ , где $n$ - количество витков на единичной длине соленоида, $I$ - ток, протекающий через соленоид. Так как соленоид является длинным, то циркуляция магнитного поля пол замкнутому контуру равна нулю, так как вектор $\overrightarrow{H}$ перпендикулярен контуру. Поэтому в середине интеграл по контуру равен нулю: вектора $H2\pi R=nI$, где $R$ - радиус соленоида. Отсюда  $H = {nI \over 2\pi R}$.

Для нахождения напряженности магнитного поля внутри тороида можем воспользоваться тем же уравнением, так как циркуляция магнитного поля по любому замкнутому контуру внутри тороида также равна нулю. Получаем $H2\pi R=nI$. Отсюда  $H = {nI \over 2\pi R}$.

## Предпроцессинг данных и подключение библиотек.

Код подключает необходимые библиотеки и проводит обработку данных из таблиц с данными, полученными при эксперименте.

```python
from csv import *
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def file_read(path):
    with open(path, 'r') as file:
        data = file.readlines()
    data[0] = data[0][3:]
    #print(data)
    #data = data[1:]
    for i in range(len(data)):
        data[i] = data[i].replace(',','.')
        data[i] = data[i].split(';')
        for j in range(len(data[i])):
            data[i][j] = float(data[i][j])
    #print(data)
    return data
```

## Задание 1. Исследование зависимости индукции магнитного поля на оси кольцевой катушки от расстояния $x$ от ее центра $(B=f(x))$ при различных значениях силы тока в катушке.

Считывание данных из файла:

```python
table1 = file_read("33.1_1.csv")
table1 = np.array(table1)
table1 = table1.transpose()
table1 = table1.tolist()
```

Считаем средние значения $<m>$ для каждого $x$ и вычисляем значения индукции магнитного поля по формуле $B=ak<m>$:

```python
a = 2.04
k = 0.42*10**(-5)
m_aver = []
for i in range(len(table1[0])):
    m = 0
    for j in range(2,6):
        m += table1[j][i]
    m /= 4
    m_aver.append(m)
table1.append(m_aver)
B = []
for i in range(len(table1[0])):
    b = round(a*k*table1[6][i],5)
    B.append(b)
table1.append(B)
```

Делаем таблицу:

```python
lines1 = {"Сила тока в кольцев ой катушке, (мА)":table1[0],\
         "Значение x, (см)":table1[1],\
         "Отброс m «зайчика» гальванометра влево 1, (дел.)":table1[2],\
         "Отброс m «зайчика» гальванометра влево 2, (дел.)":table1[3],\
         "Отброс m «зайчика» гальванометра вправо 1, (дел.)":table1[4],\
         "Отброс m «зайчика» гальванометра вправо 2, (дел.)":table1[5],\
         "Среднее значение отброса, <m> (дел.)":table1[6],\
         "Индукция магнитного поля, (Тл)":table1[7]}
t1 = pd.DataFrame(lines1)
t1
```

Считаем теоретические значения индукции магнитного поля $B$ по формуле $B=\frac{\mu_0IN}{2r_0}\sin^3{\beta}$:

```python
mu_0 = 4*np.pi*10**(-7)
r_0 = 0.177
N = 560

B_t = []
for i in range(len(table1[0])):
    beta = np.arctan(100*r_0/table1[1][i])
    B_t.append(round(mu_0*table1[0][i]/1000*N*(np.sin(beta))**3/2/r_0,5))
```

Строим графики для каждого $I$,$I'$,$I''$ (синий график - экспериментальные значения, красный - теоретические):

```python
plt.figure()
plt.subplot(311)
plt.xlabel('x')
plt.ylabel('B')
plt.title ('I')
plt.plot(table1[1][:6], table1[7][:6],\
         table1[1][:6], B_t[:6], 'r')

plt.subplot(312)
plt.xlabel('x')
plt.ylabel('B')
plt.title ('I\'')
plt.plot(table1[1][6:12], table1[7][6:12],\
         table1[1][6:12], B_t[6:12],'r')

plt.subplot(313)
plt.xlabel('x')
plt.ylabel('B')
plt.title ('I\'\'')
plt.plot(table1[1][12:], table1[7][12:],\
         table1[1][12:], B_t[12:], 'r')

plt.subplots_adjust(left=0.2, bottom=0.1, right=0.7, top=2, wspace=0.5, hspace=0.3)
plt.show()
```
